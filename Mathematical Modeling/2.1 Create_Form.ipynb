# 每个月中各个组分的均值建立新的表格

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
#解决中文显示问题
plt.rcParams['font.sans-serif'] = ['KaiTi'] # 指定默认字体
plt.rcParams['axes.unicode_minus'] = False # 解决保存图像是负号'-'显示为方块的问题
#图像显示不清楚
%config InlineBackend.figure_format ='svg'
path="E:/33-数学建模资料/研究生数学建模竞赛-2021/process_add.xlsx"
Sheet_Name=['PM1','PM2.5','PM10']
to_path="E:/33-数学建模资料/研究生数学建模竞赛-2021/Components_Mean.xlsx"
```

```python
#读取文件
def read_file(path,sheet_name):
    data=pd.read_excel(path,sheet_name=sheet_name)
    return data
```

```python
data1=read_file(path,Sheet_Name[0])#PM1
data2=read_file(path,Sheet_Name[1])#PM2.5
data3=read_file(path,Sheet_Name[2])#PM10
```

```python
concentration1=data1.groupby(by='YearMonth')['颗粒物浓度（ug/m3）'].mean()
concentration2=data2.groupby(by='YearMonth')['颗粒物浓度（ug/m3）'].mean()
concentration3=data3.groupby(by='YearMonth')['颗粒物浓度（ug/m3）'].mean()
```

```python
OC_1=data1.groupby(by='YearMonth')['OC(ug/m3)'].mean()
OC_2=data2.groupby(by='YearMonth')['OC(ug/m3)'].mean()
OC_3=data3.groupby(by='YearMonth')['OC(ug/m3)'].mean()
```

```python
EC_1=data1.groupby(by='YearMonth')['EC(ug/m3)'].mean()
EC_2=data2.groupby(by='YearMonth')['EC(ug/m3)'].mean()
EC_3=data3.groupby(by='YearMonth')['EC(ug/m3)'].mean()
```

```python
OC1_1=data1.groupby(by='YearMonth')['OC1(ug/m3)'].mean()
OC1_2=data2.groupby(by='YearMonth')['OC1(ug/m3)'].mean()
OC1_3=data3.groupby(by='YearMonth')['OC1(ug/m3)'].mean()
```

```python
OC2_1=data1.groupby(by='YearMonth')['OC2(ug/m3)'].mean()
OC2_2=data2.groupby(by='YearMonth')['OC2(ug/m3)'].mean()
OC2_3=data3.groupby(by='YearMonth')['OC2(ug/m3)'].mean()
```

```python
OC3_1=data1.groupby(by='YearMonth')['OC3(ug/m3)'].mean()
OC3_2=data2.groupby(by='YearMonth')['OC3(ug/m3)'].mean()
OC3_3=data3.groupby(by='YearMonth')['OC3(ug/m3)'].mean()
```

```python
OC4_1=data1.groupby(by='YearMonth')['OC4(ug/m3)'].mean()
OC4_2=data2.groupby(by='YearMonth')['OC4(ug/m3)'].mean()
OC4_3=data3.groupby(by='YearMonth')['OC4(ug/m3)'].mean()
```

```python
EC1_1=data1.groupby(by='YearMonth')['EC1(ug/m3)'].mean()
EC1_2=data2.groupby(by='YearMonth')['EC1(ug/m3)'].mean()
EC1_3=data3.groupby(by='YearMonth')['EC1(ug/m3)'].mean()
```

```python
EC2_1=data1.groupby(by='YearMonth')['EC2(ug/m3)'].mean()
EC2_2=data2.groupby(by='YearMonth')['EC2(ug/m3)'].mean()
EC2_3=data3.groupby(by='YearMonth')['EC2(ug/m3)'].mean()
```

```python
EC3_1=data1.groupby(by='YearMonth')['EC3(ug/m3)'].mean()
EC3_2=data2.groupby(by='YearMonth')['EC3(ug/m3)'].mean()
EC3_3=data3.groupby(by='YearMonth')['EC3(ug/m3)'].mean()
```

```python
OPC_1=data1.groupby(by='YearMonth')['OPC(ug/m3)'].mean()
OPC_2=data2.groupby(by='YearMonth')['OPC(ug/m3)'].mean()
OPC_3=data3.groupby(by='YearMonth')['OPC(ug/m3)'].mean()
```

```python
#PM1每月 每个组分的均值
data_PM1_Components=pd.concat([OC_1, EC_1, OC1_1, OC2_1, OC3_1, OC4_1,EC1_1,EC2_1,EC3_1,OPC_1], axis=1)
data_PM1_Components
```

```python
#PM2.5每月每个组分的均值
data_PM25_Components=pd.concat([OC_2, EC_2, OC1_2, OC2_2, OC3_2, OC4_2,EC1_2,EC2_2,EC3_2,OPC_2], axis=1)
data_PM25_Components
```

```python
#PM10每月每个组分均值
data_PM10_Components=pd.concat([OC_3, EC_3, OC1_3, OC2_3, OC3_3, OC4_3,EC1_3,EC2_3,EC3_3,OPC_3], axis=1)
data_PM10_Components
```

```python
#写入数据
writer=pd.ExcelWriter(to_path)
def to_file(data,str1):
    data.reset_index()
    data.to_excel(writer,sheet_name=str1)
```

```python
to_file(data_PM1_Components,'PM1')
to_file(data_PM25_Components,'PM2.5')
to_file(data_PM10_Components,'PM10')
```

```python
writer.save()
writer.close()
```

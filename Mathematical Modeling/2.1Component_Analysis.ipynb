```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import random
#解决中文显示问题
plt.rcParams['font.sans-serif'] = ['KaiTi'] # 指定默认字体
plt.rcParams['axes.unicode_minus'] = False # 解决保存图像是负号'-'显示为方块的问题
#图像显示不清楚
%config InlineBackend.figure_format ='svg'
Sheet_Name=['PM1','PM2.5','PM10']
path="E:/33-数学建模资料/研究生数学建模竞赛-2021/Components_Mean.xlsx"
```

```python
#读取文件
def read_file(path,sheet_name):
    data=pd.read_excel(path,sheet_name=sheet_name)
    return data
```

```python
data1=read_file(path,Sheet_Name[0])#PM1
data2=read_file(path,Sheet_Name[1])#PM2.5
data3=read_file(path,Sheet_Name[2])#PM10
```

# PM1中各组分时空分布特征

```python
def plot_line(YearMonth,y_param,str1,color,str2,str3):
    x=YearMonth
    y=y_param
    plt.figure(figsize=(8,6))
    plt.plot(x,y,color=color)
    plt.xticks(rotation=60)
    plt.tick_params(labelsize=15) #刻度大小
    plt.xlabel('时间',size=16)
    plt.ylabel(str1+'_'+str2,size=16)
    plt.savefig('images/Component_Analysis/'+str1+'_'+str3+'.jpg',dpi=80, bbox_inches = 'tight')
    plt.show()
```

```python
x=pd.to_datetime(data1['YearMonth'],format='%Y%m')
```

```python
def process(x,data,str1,color):
    columns=['OC(ug/m3)','EC(ug/m3)','OC1(ug/m3)','OC2(ug/m3)','OC3(ug/m3)','OC4(ug/m3)','EC1(ug/m3)','EC2(ug/m3)','EC3(ug/m3)','OPC(ug/m3)']
    columns_ylabel=['OC','EC','OC1','OC2','OC3','OC4','EC1','EC2','EC3','OPC']
    c=np.random.randint(0,len(color))
    for i,j in zip(columns,columns_ylabel):
        y=data[i]
        plot_line(x,y,str1,color[c],i,j)
```

```python
colors=['palevioletred','powderblue','seagreen','firebrick','mediumpurple','brown','teal','darkgoldenrod','cyan','darkolivegreen']
```

```python
random.shuffle(colors)
process(x,data1,'PM1',colors)
```

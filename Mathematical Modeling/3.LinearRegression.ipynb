```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import random
from sklearn.model_selection import train_test_split
from sklearn import linear_model
#解决中文显示问题
plt.rcParams['font.sans-serif'] = ['KaiTi'] # 指定默认字体
plt.rcParams['axes.unicode_minus'] = False # 解决保存图像是负号'-'显示为方块的问题
#图像显示不清楚
%config InlineBackend.figure_format ='svg'
Sheet_Name=['PM1','PM2.5','PM10']
path="E:/33-数学建模资料/研究生数学建模竞赛-2021/process_add.xlsx"
```

```python
#读取文件
def read_file(path,sheet_name):
    data=pd.read_excel(path,sheet_name=sheet_name)
    return data
```

```python
data1=read_file(path,Sheet_Name[0])#PM1
data2=read_file(path,Sheet_Name[1])#PM2.5
data3=read_file(path,Sheet_Name[2])#PM10
```

```python
# 加载训练数据
def load_data(Sheet_Name):
    data=read_file(path,Sheet_Name)
    dataset=data.loc[:,['OC(ug/m3)','EC(ug/m3)','OC/EC','OC1(ug/m3)','OC2(ug/m3)','OC3(ug/m3)','OC4(ug/m3)','EC1(ug/m3)','EC2(ug/m3)','EC3(ug/m3)','OPC(ug/m3)','颗粒物浓度（ug/m3）']]
    x=dataset.iloc[:,1:12]
    y=dataset.loc[:,['颗粒物浓度（ug/m3）']]
    X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=3)
    return X_train, X_test, y_train, y_test
```

```python
X_train, X_test, y_train, y_test = load_data(Sheet_Name[0])#PM1
```

```python
#构建线性回归模型
# def train(X_train, X_test, y_train, y_test):
reg=linear_model.LinearRegression()
reg.fit(X_train,y_train)
print("Score of Train:", reg.score(X_train, y_train))
y_pred = reg.predict(X_test)
print('y_pred:',y_pred)
print('y_test:',y_test)
```

```python
# def plot();
import seaborn as sn
sn.set_style("whitegrid")
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = ['sans-serif']
plt.rcParams['font.sans-serif'] = ['SimHei']
x_=range(len(y_test))
y_true=y_test
# y_pred
plt.figure(figsize=(8,6))
plt.plot(x_,y_true,c='red',label='真实值')#真实数据
plt.plot(x_, y_pred, c='blue', alpha=0.5, marker=(9, 3, 30),label='预测值')#预测数据
plt.legend()
plt.xticks(rotation=60)
plt.tick_params(labelsize=15) #刻度大小
# plt.xlabel('时间',size=16)
plt.ylabel('PM1_颗粒物浓度（ug/m3）',size=16)
plt.savefig('images/LinerRegression/PM1_颗粒物浓度.jpg',dpi=80, bbox_inches = 'tight')
plt.show()
```

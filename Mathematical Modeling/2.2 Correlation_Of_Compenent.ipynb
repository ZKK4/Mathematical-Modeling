```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import seaborn as sns
%config InlineBackend.figure_format ='svg'
```

```python
path="E:/33-数学建模资料/研究生数学建模竞赛-2021/process_nan_anomaly.xlsx"
Sheet_Name=['PM1','PM2.5','PM10']
```

```python
columns=['OC(ug/m3)','EC(ug/m3)','OC/EC','OC1(ug/m3)','OC2(ug/m3)','OC3(ug/m3)','OC4(ug/m3)','EC1(ug/m3)','EC2(ug/m3)','EC3(ug/m3)','OPC(ug/m3)']
```

```python
def form_compenents(columns,sheet_name):
    data=pd.read_excel(path,sheet_name=sheet_name)
    compenents=pd.DataFrame()
    for i in columns:
        compenents[i]=data[i]
    return compenents
```

```python
def plot(corr,str1):
    import seaborn as sns 
    sns.set(font_scale=1.5)
    f,ax=plt.subplots(figsize=(8,8))
    sns.heatmap(corr,annot=True,linewidths=5,fmt='.1f',ax=ax)
    f.savefig('images/Component_correlation/'+str1+'_Correlation_Of_Compenent.jpg', dpi = 60,bbox_inches = 'tight')
    
    sns.set()
    sns.pairplot(corr,size = 2 ,kind ='scatter',diag_kind='kde')    
    plt.savefig('images/'+str1+'_Correlation of PM_Compenents.jpg', dpi = 180, bbox_inches = 'tight')
```

```python
for i in Sheet_Name:
    compenents=form_compenents(columns,i)
    corr=compenents.corr()
    plot(corr,i)
```

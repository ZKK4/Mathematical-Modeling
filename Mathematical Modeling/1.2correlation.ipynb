```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import seaborn as sns
%config InlineBackend.figure_format ='svg'
```

```python
consistence=pd.DataFrame()
path="E:/33-数学建模资料/研究生数学建模竞赛-2021/process_nan_anomaly.xlsx"
Sheet_Name=['PM1','PM2.5','PM10']
for i in Sheet_Name:
    data=pd.read_excel(path,sheet_name=i)
    consistence[i]=data['颗粒物浓度（ug/m3）']
```

```python
consistence
```

```python
corr=consistence.corr(method='spearman')
corr
```

```python
import seaborn as sns 
sns.set(font_scale=1.5)
f,ax=plt.subplots(figsize=(8,8))
sns.heatmap(corr,annot=True,linewidths=5,fmt='.1f',ax=ax)
f.savefig('images/1.2_Confusion Matrix of PM_Consistence.jpg', dpi = 60)
```

```python
# 每个数字特征得分布可视化
f = pd.melt(consistence, value_vars=consistence.columns)
f
```

```python
sns.set()
sns.pairplot(consistence,size = 2 ,kind ='scatter',diag_kind='kde')
plt.savefig('images/1.2_Correlation of PM_Consistence.jpg', dpi = 80, bbox_inches = 'tight')
plt.show()
```
